```python
import sympy as sp
a,b,c,d = sp.symbols("a,b,c,d")
# vector dot operation
v1=sp.Matrix([a, b])
v2=sp.Matrix([c, d])
dot_result = v1.dot(v2)
sp.init_printing(use_unicode=True)
v1, v2, dot_result
```

```python
import sympy as s
a,b,c,x,y,z = s.symbols("a,b,c,x,y,z")
#vector cross product
v1 = s.Matrix([a,b,c])
v2 = s.Matrix([x,y,z])
cross_result = v1.cross(v2)
s.init_printing(use_unicode=True)
v1, v2, cross_result
```

```python
import sympy as s
a,b,c,x,y,z = s.symbols("a,b,c,x,y,z")
v1 = s.Matrix([a,b,c])
v2 = s.Matrix([[x,y,z]])
cross_result = v1*v2
#matrix cross product
s.init_printing(use_unicode=True)
v1, v2, cross_result
```

```python
import sympy as s
a,b,c,x,y,z,d,e,f,g,h,i,x1,y1,z1,x2,y2,z2 = s.symbols("a,b,c,x,y,z,d,e,f,g,h,i,x1,y1,z1,x2,y2,z2")
v1 = s.Matrix([[a,b,c],[d,e,f],[g,h,i]])
v2 = s.Matrix([[x,y,z],[x1,y1,z1],[x2,y2,z2]])
#matrix cross product
cross_result = v1*v2
s.init_printing(use_unicode=True)
v1, v2, cross_result
```

```python
import sympy as s
a,b,c,x,y,z,d,e,f,g,h,i,x1,y1,z1,x2,y2,z2 = s.symbols("a,b,c,x,y,z,d,e,f,g,h,i,x1,y1,z1,x2,y2,z2")
v1 = s.Matrix([[a,b,c],[d,e,f],[g,h,i]])
v2 = s.Matrix([[x,y,z],[x1,y1,z1],[x2,y2,z2]])
#elementwise dot product
dot_result = v1.multiply_elementwise(v2)
s.init_printing(use_unicode=True)
v1, v2, dot_result
```
