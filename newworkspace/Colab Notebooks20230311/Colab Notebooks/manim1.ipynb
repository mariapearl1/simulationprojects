```python
from manim import *
```

```python
%%manim -qh -v WARNING thanks
class thanks(Scene):
    def replace(self, obj1, obj2):
        self.play(
            ApplyMethod(obj1.move_to, obj2),
            FadeOut(obj2),
            run_time=3)

    def construct(self):
        txt1 = Text("SUBSCRIBE").shift(LEFT*3)

        txt2 = Text("THANKS").shift(RIGHT*3)
        self.add(txt1, txt2)
        self.replace(txt1, txt2)
        self.wait(1)
```

```python
%%manim -qh -v WARNING Subscribe

class Subscribe(Scene):
    def construct(self):
        square = Square(color=WHITE)
        my_text = Text("SUBSCRIBE", font_size=48, color=BLUE_D)
        square.surround(my_text)
        self.add(square, my_text)
        self.wait(2)
```
