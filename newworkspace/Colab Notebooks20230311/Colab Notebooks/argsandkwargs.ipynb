```python
#*args can take any number of inputs
def add(*args):
    list = []
    for num in args:
        list.append(num)
    return list
    
print(add(2, 3))#here two inputs
print(add(2, 3, 5))#here 3 inputs
print(add(2, 3, 5, 7))#here 4 inputs
print(add(2, 3, 5, 7, 9))#here 5 inputs
```

```python
#*kwargs can take any number of key value pair
def add(**kwargs):
    dict = {}
    for key,value in kwargs.items():
        dict.update({key: value})
    return dict

print(add(two=2, three=3))#here two key value pairs
print(add(two=2, three=3, five=5))#here 3 key value pairs
print(add(two=2, three=3, five=5,seven=7))#here 4 key value
```
