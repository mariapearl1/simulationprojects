```python
import turtle as t
import colorsys
t.tracer(100)
t.bgcolor('black')
t.pensize(5)
h=0

for i in range(300):
   t.width(i//100+2)
   c=colorsys.hsv_to_rgb(h,0.8,1)
   t.pencolor(c)
   h+=0.006
   t.right(120)
   t.circle(-i*0.5,180)
   t.circle(i*0.5,180)
   t.circle(i*0.5,60)

t.done()
```
