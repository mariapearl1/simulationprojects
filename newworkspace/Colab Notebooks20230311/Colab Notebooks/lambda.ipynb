```python
#lambda argument(s) : expression 
susbcribe=lambda : print('Subscribe')
susbcribe()
```

```python
# lambda that accepts one argument
subscribe = lambda th : print('Subscribe ,', th)

# lambda call
subscribe('Thank You')
```

```python
# lambda that accepts two argument
addnumbers = lambda num1,num2 : print('Output ,', num1+num2)

# lambda call
addnumbers(25,35)
```

```python
lst = [1, 2, 3, 4, 5]
#lambda in filter
# Filter all elements >=3
my_list = filter(lambda x: x>=3, lst)
print(list(my_list))
```

```python
list(map(lambda n: n * 2, [1, 2, 3, 4, 5]))
```

```python
#print all names from the list
dict_a = [{'name': 'python', 'points': 10}, {'name': 'java', 'points': 10}]
list(map(lambda x : x['name'], dict_a)) # Output: ['python', 'java'] 
```

```python
#checks the key is python
list(map(lambda x : x['name'] == "python", dict_a)) # Output: [True, False]
```
